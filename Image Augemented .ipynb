# ModerateDemented TRAin

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
import os
```

```python
datagen = ImageDataGenerator(
     rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
        )
```

```python
directory='C:/Users/joseph/Desktop/TY Sem6/PBL-2/Project/Alzheimer_s Dataset/train/ModerateDemented';
imgs = os.listdir(directory);
imgs
```

```python
import cv2
for img in imgs:
    img=cv2.imread(directory+"/"+img)
    x = img_to_array(img)
    x = x.reshape((1,) + x.shape)

    i = 0
    for batch in datagen.flow(x, batch_size=1,save_to_dir='C:/Users/joseph/Desktop/TY Sem6/PBL-2/Project/AG_Alzheimer_s Dataset/train/ModerateDemented', save_prefix="modd", save_format='jpeg'):
        i+=1
        if i>140:
            break
```
